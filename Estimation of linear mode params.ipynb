```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing 
```

```python
data = pd.read_csv("weightdataset.csv", sep=';')
data.head()
```

```python
height = data['Height'] # dependent variable 
weight = data['Weight'] # independent variable

# Standardizing data variables 
weight = preprocessing.scale(weight) 
height = preprocessing.scale(height)
```

```python
plt.figure(figsize=(14,6))
plt.plot(weight,height,'.')
plt.title('Weight and height scatter ', color='red')
plt.xlabel('weight',color='blue')
plt.ylabel('height',color='blue')
plt.show()
```

```python
Epochs = 12000
learning_rate = 0.001 
n= len(weight)

theta = np.random.rand(2) #Specifying theta parameters


def gradient_descent(weight,height,theta,Epochs,learning_rate): # Gradient descent function
    
    past_cost = [] 
    past_theta = [theta]
    
    for i in range(Epochs):
        
        linear_model = theta[0] + theta[1] * weight # linear regression
        cost = 1/(2*n) * sum((linear_model - height)**2) # Cost function
        past_cost.append(cost)
        
        theta = theta - (learning_rate*(1/n)*sum(linear_model - height)) #Updating model parameters
        
        
        past_theta.append(theta)
        
    return(past_cost,past_theta)
```

```python
past_cost,past_theta = gradient_descent(weight,height,theta,Epochs,learning_rate)
theta = past_theta[-1]
```

```python
plt.figure(figsize=(15,6))
plt.title('Cost Function')
plt.xlabel('number of iterations')
plt.ylabel('cost')
plt.plot(past_cost,'r-')
plt.show()
```

```python
print('Gradient Descent:{:.2f},{:.2f}'.format(theta[0] , theta[1]))
```

```python
linear_model = theta[0] + theta[1] * weight

plt.figure(figsize=(20,8))
plt.plot(weight,height,'.')
plt.plot(weight,linear_model,label='Regression line: height=_0 + _1*weight')
plt.xlabel('weights')
plt.ylabel('heights')
plt.title('Model parameters={:.2f},{:.2f}'.format(theta[0],theta[1]))
plt.legend(loc=4)
plt.show()
```
